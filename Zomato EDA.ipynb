```python
df1_country_code=pd.read_excel('Country-Code.xlsx')
df1_country_code.head()
```

```python
final_df=pd.merge(df1,df1_country_code,on='Country Code',how='left')
final_df.head()
```

```python
final_df.columns
```

```python
final_df.dtypes
```

```python
Country_names=final_df.Country.value_counts().index
```

```python
Country_values=final_df.Country.value_counts().values
```

```python
## Pie Chart 
plt.pie(Country_values,labels=Country_names)
```

```python
## Pie Chart top 3 Countries
plt.pie(Country_values[:3],labels=Country_names[:3],autopct='%1.2f%%')
```

Obsevation: Zomato max transactions done in India after United States than United Kingdom

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Counts'})
```

```python
ratings
```

# # Observation
1.when Rating is between 4.5 to 4.9--->Excellent
2.when Rating is between 4.0 to 3.4--->Very Good
3.when Rating is between 3.5 to 9.9--->Good
4.when Rating is between 2.5 to 3.4--->Average
5.when Rating is between 1.8 to 2.4--->Poor
5.when Rating is 0.0--->Not Rated

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Counts", data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Counts",hue='Rating color', data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
